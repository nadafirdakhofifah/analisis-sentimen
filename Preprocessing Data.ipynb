```python
import pandas as pd
import re

# Fungsi untuk membersihkan teks
def clean_text(text):
    if isinstance(text, str):  # Pastikan teks berupa string
        # Hapus URL
        text = re.sub(r'http\S+|www\S+|https\S+', '', text)

        # Hapus tag HTML
        text = re.sub(r'<.*?>', '', text)

        # Hapus emoji menggunakan regex
        text = re.sub(r'[^\x00-\x7F]+', '', text)  # Menghapus karakter non-ASCII (termasuk emoji)

        # Hapus simbol dan angka, termasuk '@' dan username seperti @matcha123
        text = re.sub(r'@[\w]+', '', text)  # Hapus @ diikuti dengan alfanumerik

        # Hapus karakter non-huruf lainnya, kecuali spasi
        text = re.sub(r'[^A-Za-z\s]', '', text)

    return text

# Membaca file CSV (misal 'data.csv' yang berisi kolom 'text')
df = pd.read_csv('boycott.csv')  # Ganti dengan nama file CSV Anda

# Pastikan kolom yang berisi teks adalah 'text'
df['cleaned_text'] = df['full_text'].apply(clean_text)

# Rename kolom
df.rename(columns={
    'full_text': 'Sebelum Cleansing',
    'cleaned_text': 'Setelah Cleansing'
}, inplace=True)

# Tampilkan tabel (tanpa print)
df[['Sebelum Cleansing', 'Setelah Cleansing']]

# Simpan hasil ke file CSV baru
df.to_csv('cleaned_boycott.csv', index=False)

print("Data telah dibersihkan dan disimpan di 'cleaned_boycott.csv'")
```

```python
import pandas as pd
import re

# Fungsi untuk case folding
def case_folding(text):
    return text.lower()

# Membaca file CSV (misal 'data.csv' yang berisi kolom 'text')
df = pd.read_csv('cleaned_boycott.csv')  # Ganti dengan nama file CSV Anda

# Pastikan kolom yang berisi teks adalah 'text'
df['case_folding'] = df['Setelah Cleansing'].apply(case_folding)

# Rename kolom
df.rename(columns={
    'Setelah Cleansing': 'Sebelum Case Folding',
    'case_folding': 'Setelah Case Folding'
}, inplace=True)

# Tampilkan tabel (tanpa print)
df[['Sebelum Case Folding', 'Setelah Case Folding']]

# Simpan hasil ke file CSV baru
df.to_csv('case folding_boycott.csv', index=False)

print("Data telah diedit dan disimpan di 'case folding_boycott.csv'")
```

```python
import pandas as pd
import re

# Membaca file CSV hasil case folding
df = pd.read_csv('case folding_boycott.csv')

# Fungsi tokenisasi menggunakan regex
def tokenization(text):
    tokens = [t for t in re.split(r'\W+', str(text)) if t]
    return tokens

# Terapkan tokenisasi ke kolom 'Setelah Case Folding'
df['Setelah Tokenization'] = df['Setelah Case Folding'].apply(tokenization)

# Rename kolom untuk kejelasan (opsional)
df.rename(columns={
    'Setelah Case Folding': 'Sebelum Tokenization'
}, inplace=True)

# Tampilkan hasil
df[['Sebelum Tokenization', 'Setelah Tokenization']]

# Simpan hasil ke file CSV baru
df.to_csv('tokenization_boycott.csv', index=False)

print("Data telah diedit dan disimpan di 'tokenization_boycott.csv'")
```

```python
import pandas as pd
import nltk
import re

# Download stopwords jika belum ada
nltk.download('stopwords')

# Ambil daftar stopwords bahasa Inggris
stopwords = nltk.corpus.stopwords.words('indonesian')

# Membaca file CSV yang sudah berisi hasil tokenisasi
df = pd.read_csv('tokenization_boycott.csv')

# Pastikan kolom tokenisasi sudah berupa list, bukan string
# Jika kolom 'Setelah Tokenization' disimpan sebagai string (misalnya: "['ak', 'prnh']"),
# kita harus konversi ke list dengan `ast.literal_eval`
import ast
df['Setelah Tokenization'] = df['Setelah Tokenization'].apply(ast.literal_eval)

# Fungsi untuk menghapus stopwords
def remove_stopwords(text):
    return [word for word in text if word.lower() not in stopwords]

# Terapkan ke kolom 'Setelah Tokenization'
df['Setelah Stopword Removal'] = df['Setelah Tokenization'].apply(remove_stopwords)

# Rename kolom untuk kejelasan (opsional)
df.rename(columns={
    'Setelah Tokenization': 'Sebelum Stopword Removal'
}, inplace=True)

# Tampilkan hasil
df[['Sebelum Stopword Removal', 'Setelah Stopword Removal']]

# Simpan hasil ke file CSV baru
df.to_csv('stopword removal_boycott.csv', index=False)

print("Data telah diedit dan disimpan di 'stopword removal_boycott.csv'")
```

```python
import pandas as pd
import ast
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Membaca file CSV hasil stopword removal
df = pd.read_csv('stopword removal_boycott.csv')

# Konversi string list ke list Python jika perlu
df['Setelah Stopword Removal'] = df['Setelah Stopword Removal'].apply(ast.literal_eval)

# Inisialisasi stemmer Bahasa Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk stemming Bahasa Indonesia
def stemming_indonesia(text):
    return [stemmer.stem(word) for word in text]

# Terapkan stemming
df['Setelah Stemming'] = df['Setelah Stopword Removal'].apply(stemming_indonesia)

# Rename kolom untuk kejelasan
df.rename(columns={
    'Setelah Stopword Removal': 'Sebelum Stemming'
}, inplace=True)

# Tampilkan hasil
df[['Sebelum Stemming', 'Setelah Stemming']]

# Simpan hasil ke file CSV baru
df.to_csv('stemming_boycott.csv', index=False)

print("Data telah diedit dan disimpan di 'stemming_boycott.csv'")
```
