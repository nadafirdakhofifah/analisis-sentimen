```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('stemming_boycott (2).csv')

# Function to label tweets based on 'Setelah Stemming'
def label_tweet(stemming_text):
    # Define keywords for labeling
    pro_keywords = ['dukungan', 'support', 'bela', 'tolong', 'boikot', 'afiliasi', 'pls', 'freepalestine', 'miris', 'sakit', 'israel','please', 'tidak', 'jangan', 'bukan', 'palestine', 'mendukung', 'keepboycotting', 'woww', 'membeli', 'ga ngeup', 'biasakan', 'bye','beli milo', 'dibeli' ]
    kontra_keywords = ['tidak', 'gak', 'ngga', 'support', 'pake', 'gak mau', 'susah']

    # Convert text to lowercase for comparison
    text = stemming_text.lower()

    # Check for pro keywords
    if any(keyword in text for keyword in pro_keywords):
        return 'pro'
    # Check for kontra keywords
    elif any(keyword in text for keyword in kontra_keywords):
        return 'kontra'
    else:
        return 'netral'

# Apply the labeling function to the 'Setelah Stemming' column
df['Setelah Pelabelan'] = df['Setelah Stemming'].apply(label_tweet)

# Rename column for clarity
df.rename(columns={
    'Setelah Stemming': 'Sebelum Pelabelan'
}, inplace=True)

# Display the results
df[['Sebelum Pelabelan', 'Setelah Pelabelan']]

# Simpan hasil ke file CSV baru
df.to_csv('labeled_boycott.csv', index=False)

print("Data telah diedit dan disimpan di 'labeled_boycott.csv'")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Memuat data dari file CSV
df = pd.read_csv('labeled_boycott.csv')  # Ganti 'data.csv' dengan nama file CSV yang sesuai

# Asumsi file CSV memiliki dua kolom: 'text' dan 'label'
documents = df['Sebelum Tokenization'].values
labels = df['Setelah Pelabelan'].values

# Ekstraksi fitur dengan TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

# Split data menjadi 80% training dan 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)

# Cek hasil
print("Jumlah data latih:", X_train.shape[0])
print("Jumlah data uji:", X_test.shape[0])
```
