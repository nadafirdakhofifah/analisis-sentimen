```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Baca data dari CSV
data = pd.read_csv('labeled_boycott (10).csv')

# 2. Ambil kolom teks
texts = data['Sebelum Tokenization']

# 3. TF-IDF vectorization
vectorizer = TfidfVectorizer(max_features=1000)  # batasi jumlah fitur jika ingin eksak 1000
X = vectorizer.fit_transform(texts)

# 4. Tampilkan shape dari matriks TF-IDF
print("Shape matriks TF-IDF:", X.shape)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Membaca data CSV
df = pd.read_csv('labeled_boycott (10).csv')  # Ganti 'data.csv' dengan lokasi file CSV Anda

# Misalkan kolom yang berisi teks adalah 'Sebelum Pelabelan'
documents = df['Sebelum Pelabelan'].tolist()

# Inisialisasi TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Ekstraksi fitur dari data teks
tfidf_matrix = vectorizer.fit_transform(documents)

# Menyimpan hasil matriks TF-IDF ke dalam df
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Menampilkan hasil dalam bentuk DataFrame (df)
print("Matriks TF-IDF:")
print(df_tfidf)

# Menampilkan fitur (kata-kata unik)
print("\nFitur (kata-kata):")
print(vectorizer.get_feature_names_out())

# Menyimpan hasil TF-IDF dalam CSV
df_tfidf.to_csv('TF-IDF_boycott.csv', index=False)
print("Data berhasil disimpan dalam 'TF-IDF_boycott.csv'.")
```
